```r
# Definimos una funcion escalÃ³n
hardlim <- function(x) {
    y <- ifelse(x>0,1,0)
        return(y)
}


# Definimos una neurona x=dato w=peso b=intercepto
neurona <- function(x,w,b){
    salida_n <- hardlim(crossprod(x,w)+b)
}


#Definimos funcion de prueba del perceptron
TestPerceptron <- function(X,W,b){
    yp<- numeric(dim(X)[1])
    for(i in 1:dim(X)[1])
        yp[i] = neurona(X[i,],W,b)
    return(yp)
}      
```

```r
# APRENDIZAJE DE LA NEURONA
# funcion de entrenamiento

entrenamiento <- function(X,Y){
    
    # inicializamos
    W<- c(1.0,-0.8)
    b<-0 
    
    # Probando el perceptron por primera vez
    TestPerceptron(X,W,b)
    
    # siempre que la prediccion sea diferente de los datos
    while(!identical(yp,Y)){
        
        # Variables de intercambio (switch)
        yp_sw <- yp[((Y-yp) != 0)]
        y_sw <- Y[((Y-yp) != 0)]
        X_sw <- X[((Y-yp) != 0),]
        # la expresion yp[((Y-yp) != 0)]  encuentra elementos con errores de prediccion 
        # restando los vectores Y-yp y comparando != 0 (regresa T/F)
        
        # Actualizando pesos... utilizamos el primero de los errores encontrados
        W <- W + (y_sw[1] - yp_sw[1]) * X_sw[1]
        b <- b + (y_sw[1] - yp_sw[1])
        print("pesos actualizados")
        
        # Probando el perceptron de nuevo
        TestPerceptron(X,W,b)
    }
    
    return(c(W,b,yp))
}
```

```r
# Probamos la funcion de entrenamiento
X<-cbind(c(1,-1,0),c(2,2,-1))
Y<-c(1,0,0)

a<-entrenamiento(X,Y)
a
```

```r
# Entrenamiento PASO a PASO

# Recalculando pesos 
#recordamos los datos del ejemplo (X1,X2)-> Y
X<-cbind(c(1,-1,0),c(2,2,-1))
Y<-c(1,0,0)

# Les asignamos un peso aleatorio a X1 y X2... suponemos un umbral de 0
W<- c(1.0,-0.8)
b<-0 

##############################################################################################################
sprintf("Prediccion inicial")

# Llamamos a la neurona y le presentamos el primer par de datos, los pesos y el intercepto aleatorios
yp = neurona(X[1,],W,b)
sprintf("Salida= %d , Prediccion= %d",Y[1],yp[1])    # salida predecida... no es correcta

################################################################################################################
sprintf("Reajustamos pesos usando observacion 1")

# Corregimos los pesos con la funcion de ajuste W(new) = W(old) + error*X   donde error = salida-salida predecida
W<-W + (Y[1]-yp[1])*X[1,]
sprintf("Nuevos pesos")
W
# Corregimos el umbral con la funcion de ajuste b(new) = b(old) + error
b<-b+ (Y[1]-yp);
sprintf("Nuevo Umbral")
b[1]

sprintf("Segunda prediccion")
# Presentamos segundo par de datos y los nuevos pesos y umbral
yp = neurona(X[1,],W,b)
sprintf("Salida= %d , Prediccion= %d",Y[1],yp[1])
yp = neurona(X[2,],W,b)
sprintf("Salida= %d , Prediccion= %d",Y[2],yp[1])   # salida predecida... no es correcta

#graficando de nuevo
ggplot(datos, aes(x=X1, y=X2, color=as.factor(Y)))+
       geom_point(size=5)+
       geom_abline(intercept=(-b/W[2]), slope=(-W[1]/W[2]))

#################################################################################################################
sprintf("Nuevamente reajustamos pesos usando la observacion 2")
# Corregimos los pesos con la funcion de ajuste W(new) = W(old) + error*X   donde error = salida-salida predecida
W<-W + (Y[2]-yp[1])*X[2,]
sprintf("Nuevos pesos")
W
# Corregimos el umbral con la funcion de ajuste b(new) = b(old) + error
b<-b+ (Y[2]-yp);
sprintf("Nuevo Umbral")
b[1]

sprintf("Tercer prediccion")
# Presentamos Tercer par de datos + pesos y umbral nuevos
yp = neurona(X[1,],W,b)
sprintf("Salida= %d , Prediccion= %d",Y[1],yp[1])
yp = neurona(X[2,],W,b)
sprintf("Salida= %d , Prediccion= %d",Y[2],yp[1])  
yp = neurona(X[3,],W,b)
sprintf("Salida= %d , Prediccion= %d",Y[3],yp[1])   # salida predecida... no es correcta

#graficando de nuevo
ggplot(datos, aes(x=X1, y=X2, color=as.factor(Y)))+
       geom_point(size=5)+
       geom_abline(intercept=(-b/W[2]), slope=(-W[1]/W[2]))

#################################################################################################################
sprintf("Nuevamente reajustamos pesos usando la observacion 3")

# Corregimos los pesos con la funcion de ajuste W(new) = W(old) + error*X   donde error = salida-salida predecida
W<-W + (Y[3]-yp[1])*X[3,]
sprintf("Nuevos pesos")
W
# Corregimos el umbral con la funcion de ajuste b(new) = b(old) + error
b<-b+ (Y[3]-yp);
sprintf("Nuevo Umbral")
b[1]

sprintf("Ultima prediccion")

yp = neurona(X[1,],W,b)
sprintf("Salida= %d , Prediccion= %d",Y[1],yp[1])
yp = neurona(X[2,],W,b)
sprintf("Salida= %d , Prediccion= %d",Y[2],yp[1])  
yp = neurona(X[3,],W,b)
sprintf("Salida= %d , Prediccion= %d",Y[3],yp[1])

#graficando de nuevo
ggplot(datos, aes(x=X1, y=X2, color=as.factor(Y)))+
       geom_point(size=5)+
       geom_abline(intercept=(-b/W[2]), slope=(-W[1]/W[2]))
```
