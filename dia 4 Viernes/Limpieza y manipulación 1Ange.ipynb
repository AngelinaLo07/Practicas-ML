### purr

usando la función map obten los años bisiestos del siglo XX

```r
# Funcion año bisiesto: en archivo RepasodeR
leap <- function(y){
    return (y%%4 ==0 && y%%100!=0 || y%%400==0)}

# map es semejante a apply... aplica una funcion a todos los elementos de un vector

year <- 1901:2000 #Vector de años del sigo XX
bis <- map(year, leap) #Aplicar la funcion Leap a todos los años (Regresa un vector T/F)

bis_XX<- year[unlist(bis)] #obtener los datos en el vector year
bis_XX
```

### tidyr

```r
casos_diarios <-  read.csv(file = "Casos_Diarios_Estado_Nacional.csv")
str(casos_diarios) #muestra una lista de variables contenidas en el objeto
```

```r
# Cantidad de columas que tiene el objeto (856) NOTA: objeto en formato ancho: cada columna es una variable
length(names(casos_diarios)) 

# Convertir a formato Largo donde tenemos solo una columna de datos(value) y otras 2 como identificadores(ID,key)
# gather(data, key=variable de procedencia, value=valor de la variable)
casos_diario_list <- gather(data=casos_diarios, key ="dias", value="infectados", 4:856)
casos_diario_list

# NOTA: 
#   cve_ent, poblacion y nombre son identificadores
#   dias sera nuestro key
#   numero de infectados será nuestra variable
```

```r
#Funcion para eliminar la X en el vector de "dias"
normalizarfecha <- function(fecha){
return(str_replace_all(fecha,"X","")  %>% str_replace_all("\\.","/"))
}

#OTRAS FORMAS DE QUITAR LA X AL VECTOR DE dias ############################################
#a <- casos_diario_list[4] %>%  str_replace_all("X","")  %>% str_replace_all("\\.","/"); a
#vect <- str_replace_all(casos_diario_list[4],"X","")
#str_replace_all(vect,"\\.","/")
###########################################################################################

fecha = map(casos_diario_list[4],normalizarfecha); #aplicamos la funcion normalizarfecha

# Quitamos el vector "dias" y Anadimos el nuevo vector "fecha" 
casos_diarios_fecha <- cbind(casos_diario_list[c(TRUE,TRUE,TRUE,FALSE,TRUE)],fecha)
#Reordenamos 
casos_diarios_fecha <- casos_diarios_fecha[c(1,2,3,5,4)]
#Separamos el vector "fecha" (que se sigue llamando dias) en 3 nuevos vectores "dia" "mes" "año"
datos <- separate( data = casos_diarios_fecha,
                   col =dias,
                   into = c("dia","mes","año"),
                   sep = '/')

datos
```

### forcats

```r
# funcion para codificar los meses 
cod_Meses <- function(x) {
labels_meses <- c(
    "Ene","Feb","Mar","Abr","May","Jun",
    "Jul","Ago","Sep","Oct","Nov","Dic")
    return(factor(x,levels = (1:12),labels = labels_meses))}

datos$mes %>%
    unlist() %>%
    as.integer()%>%
    cod_Meses()



#fct_recode(cod_Meses(as.integer(unlist(salida[5]))), "Enero" = "Ene", "Febrero" = "Feb")
#fct_expand(cod_Meses(as.integer(unlist(salida[5]))), "Otro")
#fct_collapse(cod_Meses(as.integer(unlist(salida[5]))), "1trimestre" =c("Ene","Feb","Mar"))

```
