### Listas

```r
A <- list("TRUE","a","7",27);A

up_cast <-function(x){
    suppressWarnings({ #Global No Warnings
    try({ 
        if(!is.na(as.logical(x))) #ask if X is different form NA when try to transform on logical
            return(as.logical(x)) #if yes return
    },silent = TRUE) #error report supressed 

    
    try({
        if(!is.na(as.integer(x)))     
            return(as.integer(x)) 
    },silent = TRUE)

    
    try({
        if(!is.na(as.double(x)))     
            return(as.double(x)) 
    },silent = TRUE)
    })
    
  return(x)
}

lista <- lapply(A,up_cast)  # consider using map fuction
str(lista)
```
